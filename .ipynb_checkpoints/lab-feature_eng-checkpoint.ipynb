There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
df_space = spaceship.dropna()
df_space.isnull().sum()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
df_space["Cabin"].value_counts()

cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T']

def replace_cabin(cabin):
    if cabin[0] == cabin_list[0]:
        return cabin_list[0]
    elif cabin[0] == cabin_list[1]:
        return cabin_list[1]
    elif cabin[0] == cabin_list[2]:
        return cabin_list[2]
    elif cabin[0] == cabin_list[3]:
        return cabin_list[3]
    elif cabin[0] == cabin_list[4]:
        return cabin_list[4]
    elif cabin[0] == cabin_list[5]:
        return cabin_list[5]
    elif cabin[0] == cabin_list[6]:
        return cabin_list[6]
    elif cabin[0] == cabin_list[7]:
        return cabin_list[7]
    elif cabin[0] == cabin_list[8]:
        return cabin_list[8]
    return cabin 

df_space.loc[:, "Cabin"] = df_space["Cabin"].apply(replace_cabin)
```
